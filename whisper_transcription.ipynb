## Step 1: Setting Up the Environment

```python
"""
# W2D2_LAB_Whisper_STT_Implementation
- Week 2 / Day 2
- Student: Andreas Papachristophorou
- Course: AI Consulting & Integration 2026-07
- Date: 2026-07-14
"""

import os
import json
from pathlib import Path
from openai import OpenAI
from dotenv import load_dotenv
from pydub import AudioSegment

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("OPENAI_API_KEY is not set")

# Create the OpenAI client so we can call Whisper.
client = OpenAI(api_key=api_key)


# Create folders for input and output files.
AUDIO_DIR = Path("audio")
OUTPUT_DIR = Path("output")
AUDIO_DIR.mkdir(exist_ok=True)
OUTPUT_DIR.mkdir(exist_ok=True)
```

## Step 3: Basic Transcription (Without Chunking)

```python
# This function sends one MP3 file to Whisper and returns the text.
def transcribe_basic(audio_path):
    with open(audio_path, "rb") as file:
        transcript = client.audio.transcriptions.create(
            model="whisper-1",
            file=file
        )
    return transcript.text

# Run a basic transcription.
basic_text = transcribe_basic(AUDIO_FILE)
print("Basic transcription:")
print("-" * 50)
print(basic_text)
```

## Step 6: Implementing Audio Chunking

```python
# Split the MP3 into smaller pieces.
# This is useful when the file is long or too large.
def chunk_audio(audio_path, chunk_seconds=10):
    audio = AudioSegment.from_mp3(audio_path)
    chunk_ms = chunk_seconds * 1000
    
    chunks = []
    
    for i, start in enumerate(range(0, len(audio), chunk_ms), start=1):
        chunk = audio[start:start + chunk_ms]
        chunk_path = OUTPUT_DIR / f"ME025_chunk_{i:03d}.mp3"
        chunk.export(chunk_path, format="mp3")
        
        chunks.append({
            "path": chunk_path,
            "offset_seconds": start / 1000
        })
    
    return chunks


chunks = chunk_audio(AUDIO_FILE, chunk_seconds=10)
print("Number of chunks created:", len(chunks))

for chunk in chunks:
    print(chunk["path"], "offset:", chunk["offset_seconds"], "seconds")
```

## Step 7: Transcribing Chunks with Timestamps.

```python
# Convert seconds into SRT timestamp format.
def seconds_to_srt_time(seconds):
    milliseconds = int(round(seconds * 1000))
    hours, milliseconds = divmod(milliseconds, 3600000)
    minutes, milliseconds = divmod(milliseconds, 60000)
    secs, milliseconds = divmod(milliseconds, 1000)
    return f"{hours:02d}:{minutes:02d}:{secs:02d},{milliseconds:03d}"

# Transcribe one chunk and get timestamped output.
def transcribe_chunk_verbose(chunk_path, prompt_text=None):
    with open(chunk_path, "rb") as file:
        kwargs = {
            "model": "whisper-1",
            "file": file,
            "response_format": "verbose_json",
            "timestamp_granularities": ["segment"]
        }
        if prompt_text:
            kwargs["prompt"] = prompt_text
        
        transcript = client.audio.transcriptions.create(**kwargs)
    return transcript

# Transcribe all chunks and adjust timestamps by the chunk offset.
def transcribe_all_chunks(chunks, prompt_text=None):
    all_segments = []
    
    for chunk in chunks:
        result = transcribe_chunk_verbose(chunk["path"], prompt_text=prompt_text)
        offset = chunk["offset_seconds"]
        
        for seg in result.segments:
            all_segments.append({
                "start": seg.start + offset,
                "end": seg.end + offset,
                "text": seg.text.strip()
            })
    
    return all_segments

chunk_segments = transcribe_all_chunks(chunks)
print("Total segments from chunks:", len(chunk_segments))

for seg in chunk_segments[:5]:
    print(seconds_to_srt_time(seg["start"]), "-->", seconds_to_srt_time(seg["end"]), seg["text"])
```

## Step 8: Export the result into readable files.

```python
# Save plain text transcript.
def save_txt(segments, output_path):
    with open(output_path, "w", encoding="utf-8") as f:
        for seg in segments:
            f.write(f'[{seconds_to_srt_time(seg["start"])} --> {seconds_to_srt_time(seg["end"])}] {seg["text"]}\n')


# Save JSON transcript.
def save_json(segments, output_path):
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(segments, f, indent=2, ensure_ascii=False)


# Save SRT transcript.
def save_srt(segments, output_path):
    with open(output_path, "w", encoding="utf-8") as f:
        for i, seg in enumerate(segments, start=1):
            f.write(f"{i}\n")
            f.write(f'{seconds_to_srt_time(seg["start"])} --> {seconds_to_srt_time(seg["end"])}\n')
            f.write(f'{seg["text"]}\n\n')


# Export files.
save_txt(chunk_segments, OUTPUT_DIR / "ME025_transcript.txt")
save_json(chunk_segments, OUTPUT_DIR / "ME025_transcript.json")
save_srt(chunk_segments, OUTPUT_DIR / "ME025_transcript.srt")

print("Export complete.")
print("TXT saved:", OUTPUT_DIR / "ME025_transcript.txt")
print("JSON saved:", OUTPUT_DIR / "ME025_transcript.json")
print("SRT saved:", OUTPUT_DIR / "ME025_transcript.srt")

print("Export complete.")
```
